```python
nnls_brier_l = []; hillclimb_brier_l = []
for i in range(100) : 
    
    # read ith dataset
    train_X_dir = '/Users/pio/Google 드라이브/github/survival ensemble/dataset/pbc2' + '_seed_' + str(i) + '_train_stack_X.csv'
    train_y_dir = '/Users/pio/Google 드라이브/github/survival ensemble/dataset/pbc2' + '_seed_' + str(i) + '_train_stack_y.csv'
    train_w_dir = '/Users/pio/Google 드라이브/github/survival ensemble/dataset/pbc2' + '_seed_' + str(i) + '_train_stack_w.csv'
    test_X_dir = '/Users/pio/Google 드라이브/github/survival ensemble/dataset/pbc2' + '_seed_' + str(i) + '_test_stack_X.csv'
    test_y_dir = '/Users/pio/Google 드라이브/github/survival ensemble/dataset/pbc2' + '_seed_' + str(i) + '_test_stack_y.csv'

    train_X = pd.read_csv(train_X_dir)
    train_y = pd.read_csv(train_y_dir)
    train_w = np.array(pd.read_csv(train_w_dir)).reshape(-1,)
    test_X = pd.read_csv(test_X_dir)
    test_y = pd.read_csv(test_y_dir)
    
    # fitting Superlearner
    nnls = nnls_constraint()
    nnls.fit(x = train_X, y= train_y, w =train_w )
    
    # fitting Hillclimber    
    hillclimber = hillclimb(max_iter= 2000, early_stop_n = 100, early_stop_eps = 10**(-4))
    hillclimber.fit(x= train_X, y=train_y, w=train_w)
    
    # brier score 
    nnls_brier = brier_score_loss(test_y, nnls.predict(x = test_X)) ; nnls_brier_l.append(nnls_brier_l)
    hillclimb_brier = brier_score_loss(test_y, hillclimber.predict(x = test_X)) ; hillclimb_brier_l.append(hillclimb_brier_l)

    
    print('######')
    print(i)
        
    print('% improvement : ' + str(round((nnls_brier - hillclimb_brier) / nnls_brier * 100,3)))

    
```
